```python
def clean_function(df):
    df.drop(["Unnamed: 0"], axis = 1, inplace= True)
    df.drop(["Length of Stay"], axis = 1, inplace = True)
    #drop non-necesary event
    df.drop(df[(df.Event !=  "Move In") & (df.Event !=  "Move Out") & (df.Event !=  "Rent Change")].index, inplace=True)
    #change to datetype
    df["DateOccurred"]= pd.to_datetime(df['DateOccurred'], format='%Y-%m-%d', utc=False)
    df["DateMoveOut"]= pd.to_datetime(df['DateMoveOut'], format='%Y-%m-%d', utc=False)
    #sort by tenant and date
    df = df.sort_values(["Tenant_Code","DateOccurred"])
    #change vacany column to correspond to rentchange
    for x in range(len(df)):
        event = df["Event"].iloc[x]
        prevevent = df["Event"].iloc[x-1]
        tenant = df["Tenant_Code"].iloc[x]
        prevtenant = df["Tenant_Code"].iloc[x-1]
        if tenant == prevtenant:
            if event == "Move Out":
                if prevevent =="Rent Change":
                    df["Vacancy"].iloc[x-1] = 1
    #calculate change in rent
    r = []
    chrent = []
    lrent = []
    for x in range(len(df)):
        tencode = df["Tenant_Code"].iloc[x]
        prevten = df["Tenant_Code"].iloc[x-1]
        if tencode == prevten:
            rent = df["Rent"].iloc[x]
            prevrent = df["Rent"].iloc[x-1]
            changerent = rent-prevrent
            chrent.append(changerent)
        else:
            chrent.append(0)
    df["chrent"]=chrent
    #add moveout date to associated rentchange
    for x in range(len(df)):
        event = df["Event"].iloc[x]
        prevevent = df["Event"].iloc[x-1]
        tenant = df["Tenant_Code"].iloc[x]
        prevtenant = df["Tenant_Code"].iloc[x-1]
        if tenant == prevtenant:
            if event == "Move Out":
                if prevevent =="Rent Change":
                    df["DateMoveOut"].iloc[x-1] = df["DateMoveOut"].iloc[x]
    #calculate length of stay
    df["LengthStay"]=df["DateMoveOut"]-df["DateOccurred"]
    df.drop(df[(df.Rent == 0)].index, inplace=True)
    df.drop(df[(df.Event == "Move Out")].index, inplace=True)
    df.drop(df[(df.Event == "Move In")].index, inplace =True)
    return df
```
