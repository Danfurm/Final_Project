```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
```

```python
apt_types = "common units.csv"
apt_types_df = pd.read_csv(apt_types, header=1)
apt_types_df = apt_types_df.drop(columns = ["Bldg_Code", "Available_Date", "Floor_Code", "Country", "Ref_Property_Id", "Ext_Ref_Unit_Id", "Ref_Building_Id", "Ref_Floor_Id", "Address_2","Address_3", "Address_4", "Location", 'Attributes_2', 'Attributes_3', 'Attributes_4', 'Attributes_5',
       'Attributes_6', 'Attributes_7', 'Attributes_8', 'Attributes_9',
       'Attributes_10', 'MLA', 'Lease_Type', 'Userdefined_1', 'Userdefined_2',
       'Userdefined_3', 'Userdefined_4', 'Userdefined_5', 'Userdefined_6',
       'Userdefined_7', 'Userdefined_8', 'Userdefined_9', 'Userdefined_10'])
apt_types_df.head()
```

```python
change = "Rent_Changed_Tenants.csv"
rentchange = pd.read_csv(change)
rentchange.reset_index(drop = True)
rentchange = rentchange.drop(["Unnamed: 0"], axis =1)
rentchange.head()
```

```python
nochange = "Tenants_nochange.csv"
norentchange = pd.read_csv(nochange)
norentchange.reset_index(drop = True)
norentchange = norentchange.drop(["Unnamed: 0"], axis =1)
norentchange.head()
```

```python
rentchange_att = pd.merge(rentchange, apt_types_df, on = "Unit_Code", how = "inner" )
rentchange_att = rentchange_att.drop(columns = ["Property_Code_y", "Date_Ready", "Address_1", "Rent_y", "Rent_Ready", "Rental_Type", "Exclude", "Notes", "Unit_Type"])
rentchange_att = rentchange_att.rename(columns = {"Property_Code_x": "Property_Code", "Rent_x": "Rent", "Attributes_1": "Bathroom"})
```

```python
rentchange_att = rentchange_att.loc[rentchange_att["Event"].isin(["Rent Change", "Move Out"])]
rentchange_att.to_csv("rentchange_attributes.csv")
rentchange_att.head()
```

```python
nochange_att = pd.merge(norentchange, apt_types_df, on = "Unit_Code", how = "inner" )
nochange_att = nochange_att.drop(columns = ["Property_Code_y", "Date_Ready", "Address_1", "Rent_y", "Rent_Ready", "Rental_Type", "Exclude", "Notes", "Unit_Type"])
nochange_att = nochange_att.rename(columns = {"Property_Code_x": "Property_Code", "Rent_x": "Rent", "Attributes_1": "Bathroom"})
nochange_att.head()
```

```python
nochange_att = nochange_att.loc[nochange_att["Event"].isin(["Move In", "Move Out"])]
nochange_att.to_csv("norentchange_attributes.csv")
nochange_att.head()
```
