```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
```

```python
tenants = "Tenant_History.csv"
tenants_df = pd.read_csv(tenants)
tenants_df.head()
```

```python
tenants_df = tenants_df.drop(columns = ["Ref_Property_ID", "Ext_Ref_Tenant_Id", "Ref_Unit_Id"])
```

```python
rent_change = tenants_df.loc[tenants_df["Event"] == "Rent Change"]
rent_change = rent_change.groupby("Property_Code")
rent_change.head(25)
```

```python
sh = tenants_df.loc[tenants_df["Property_Code"] == "sh"]
sh_rent = sh.loc[sh["Event"] == "Rent Change"]
sh_rent.head(25)
```

```python
#sh_rent['DateOccurred'] = sh_rent['DateOccured'].astype('datetime64[ns]')
new = sh_rent["DateOccurred"].str.split("/", n = 3, expand = True)
month =[]
day = []
year = []
for x in new[0]:
    if len(x) < 2:
        x = str("0")+str(x)
    month.append(x)
for x in new[1]:
    if len(x) < 2:
        x = str("0")+str(x)
    day.append(x)
for x in new[2]:
    year.append(x)
newdate =[]
for x in range(len(day)):
    newdate.append(year[x]+month[x]+day[x])
sh_rent["Newdate"] = newdate
sh_rent["Newdate"]
sortedsh=sh_rent.sort_values(["Unit_Code","Newdate"])
sortedsh
```

```python
r = []
chrent = []
lrent = []
for x in range(len(sortedsh)):
    unitcode = sortedsh["Unit_Code"].iloc[x]
    prevunit = sortedsh["Unit_Code"].iloc[x-1]
    if unitcode==prevunit:
        rent = sortedsh["Rent"].iloc[x]
        prevrent = sortedsh["Rent"].iloc[x-1]
        changerent = rent-prevrent
        chrent.append(changerent)
    else:
        chrent.append(0)
sortedsh["chrent"]=chrent
sortedsh
```
