```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime as dt
```

```python
tenants = "Tenant_History.csv"
tenants_df = pd.read_csv(tenants)
tenants_df.head()
```

```python
tenants_df = tenants_df.drop(columns = ["Ref_Property_ID", "Ext_Ref_Tenant_Id", "Ref_Unit_Id", "Date"])
```

```python
tenants_df["DateOccurred"] = pd.to_datetime(tenants_df["DateOccurred"], format= '%m/%d/%Y')
tenants_df["DateApply"] = pd.to_datetime(tenants_df["DateApply"], format= '%m/%d/%Y')
tenants_df["DateApprove"] = pd.to_datetime(tenants_df["DateApprove"], format= '%m/%d/%Y')
tenants_df["DateSign"] = pd.to_datetime(tenants_df["DateSign"], format= '%m/%d/%Y')
tenants_df["DateMoveIn"] = pd.to_datetime(tenants_df["DateMoveIn"], format= '%m/%d/%Y')
tenants_df["DateMoveOut"] = pd.to_datetime(tenants_df["DateMoveOut"], format= '%m/%d/%Y')
tenants_df["DateLeaseFrom"] = pd.to_datetime(tenants_df["DateLeaseFrom"], format= '%m/%d/%Y')
tenants_df["DateLeaseTo"] = pd.to_datetime(tenants_df["DateLeaseTo"], format= '%m/%d/%Y')
tenants_df["DateNotice"] = pd.to_datetime(tenants_df["DateNotice"], format= '%m/%d/%Y')
tenants_df["Length of Stay"]= tenants_df["DateMoveOut"] - tenants_df["DateMoveIn"]
tenants_df.sort_values(by=["Unit_Code"])
```

```python
occupancy = tenants_df.sort_values(["DateOccurred", "Tenant_Code"])
occupancy = occupancy.drop(columns = ["DateApply", "DateApprove", "DateSign", "DateLeaseFrom", "DateLeaseTo", "DateNotice", "RentedEvent", "UnRentedEvent", "Status", "Deposit", "Notes", "Reason", "MTM"])
filter = ["Move In", "Move Out", "Rent Change"]
occupancy = occupancy.loc[occupancy["Event"].isin(filter)]
occupancy.reset_index(drop = True)
```

```python
def vacancy (row):
    if row["Event"] == "Move Out":
        return "V"
    elif row["Event"] == "Move In":
        return "O"
    else:
        return "O"
    
occupancy ["Occupied?"] = occupancy.apply(vacancy, axis = 1) 
occupancy.head()
```

```python
year_2013 = occupancy[occupancy["DateOccurred"].dt.year == 2013]
year_2014 = occupancy[occupancy["DateOccurred"].dt.year == 2014]
year_2015 = occupancy[occupancy["DateOccurred"].dt.year == 2015]
year_2016 = occupancy[occupancy["DateOccurred"].dt.year == 2016]
year_2017 = occupancy[occupancy["DateOccurred"].dt.year == 2017]
year_2018 = occupancy[occupancy["DateOccurred"].dt.year == 2018]
idx = pd.period_range(min(year_2013.DateOccurred), max(year_2013.DateOccurred), freq = "1M")
```
